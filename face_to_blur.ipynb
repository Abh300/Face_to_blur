```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from mtcnn import MTCNN

%matplotlib inline
```

```python
pets = cv2.imread("pets.jpg")
pets = cv2.cvtColor(pets,cv2.COLOR_BGR2RGB)
plt.imshow(pets)
```

```python
mtcnn = MTCNN()
```

```python
faces = mtcnn.detect_faces(pets)
draw_rect = np.copy(pets)

for face in faces:
    x,y,w,h = face['box']
    draw_rect = cv2.rectangle(draw_rect,(x,y),(x+w, y+h), (255,255,255),2)

plt.imshow(draw_rect)
```

```python
pet_face = mtcnn.detect_faces(pets)
blurred = np.copy(pets)

for face in pet_face:
    x,y,w,h = face['box']
    blurred[y:y+h,x:x+w] = cv2.blur(blurred[y:y+h,x:x+w],(80,80))

plt.imshow(blurred)
```
